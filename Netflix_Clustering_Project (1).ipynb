```python
df_cleaned = df.dropna(subset=['type', 'rating', 'duration', 'release_year', 'country'])
df_cleaned.reset_index(drop=True, inplace=True)
```

```python
# Create a new column 'duration_int'
def convert_duration(x):
    if 'Season' in x:
        num = x.split()[0]
        return int(num) * 60  # Assume 1 Season = 60 min
    elif 'min' in x:
        return int(x.split()[0])
    else:
        return np.nan

df_cleaned.loc[:, 'duration_int'] = df_cleaned['duration'].apply(convert_duration)

```

```python
# 1. Encode type
df_cleaned = df.dropna(subset=['type', 'rating', 'duration', 'release_year', 'country']).copy()


# 2. Fill nulls in 'cast' and 'director' — without inplace
df_cleaned['cast'] = df_cleaned['cast'].fillna('')
df_cleaned['director'] = df_cleaned['director'].fillna('')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler
```

```python
df_cleaned['text_features'] = (
    df_cleaned['listed_in'].fillna('') + ' ' +
    df_cleaned['description'].fillna('') + ' ' +
    df_cleaned['cast'].fillna('') + ' ' +
    df_cleaned['director'].fillna('')
)
```

Combined genres, description, cast, and director into one feature column for text-based analysis.


```python
# Use max_features to limit dimensionality for clustering
tfidf = TfidfVectorizer(stop_words='english', max_features=1000)

# Apply TF-IDF to the combined text column
tfidf_matrix = tfidf.fit_transform(df_cleaned['text_features'])
```

```python
# Select numeric features
numeric_features = df_cleaned[['release_year', 'duration_int', 'type_encoded']].copy()

# Normalize numeric features to same scale
scaler = MinMaxScaler()
scaled_numeric = scaler.fit_transform(numeric_features)

# Combine text (TF-IDF) and numeric features
from scipy.sparse import hstack
final_features = hstack([tfidf_matrix, scaled_numeric])
```

```python
# 1. Convert duration column to int
def convert_duration(duration):
    if 'min' in duration:
        return int(duration.split()[0])
    elif 'Season' in duration:
        return int(duration.split()[0]) * 60  # 1 season ≈ 60 mins
    else:
        return 0  # unknown format

df_cleaned.loc[:, 'duration_int'] = df_cleaned['duration'].apply(convert_duration)

# 2. Encode type column
df_cleaned.loc[:, 'type_encoded'] = df_cleaned['type'].map({'Movie': 0, 'TV Show': 1})
```

```python
from sklearn.preprocessing import MinMaxScaler
from scipy.sparse import hstack

# 1. Select numeric features
numeric_features = df_cleaned[['release_year', 'duration_int', 'type_encoded']].copy()

# 2. Normalize numeric features
scaler = MinMaxScaler()
scaled_numeric = scaler.fit_transform(numeric_features)

# 3. Combine TF-IDF and numeric features
final_features = hstack([tfidf_matrix, scaled_numeric])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
wcss = []  # Within-Cluster Sum of Squares

# Try cluster counts from 2 to 10
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(final_features)
    wcss.append(kmeans.inertia_)

# Plot WCSS to find the "elbow"
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
# Final clustering with optimal k
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
df_cleaned['cluster'] = kmeans.fit_predict(final_features)
```

```python
# Check few entries from each cluster
df_cleaned[['title', 'listed_in', 'release_year', 'cluster']].sample(10)
```

```python
# Show top 5 genres in each cluster
for cluster_num in range(df_cleaned['cluster'].nunique()):
    print(f"\nCluster {cluster_num} - Top Genres:")
    genres = df_cleaned[df_cleaned['cluster'] == cluster_num]['listed_in']
    all_genres = ','.join(genres).split(',')
    top_genres = pd.Series(all_genres).value_counts().head(5)
    print(top_genres)
```

```python
# Show 5 random titles from each cluster
for cluster_num in range(df_cleaned['cluster'].nunique()):
    print(f"\n📺 Cluster {cluster_num} - Sample Titles:")
    titles = df_cleaned[df_cleaned['cluster'] == cluster_num]['title'].sample(5, random_state=42)
    print(titles.to_list())
```
